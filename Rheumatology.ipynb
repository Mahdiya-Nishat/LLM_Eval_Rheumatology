```python
print("Running Mixed Effects Model...")

# Python MixedLM handles one grouping structure naturally.
# We use Evaluator_ID as group and Domain as fixed effect.
# This is the closest stable Colab-ready approximation to your R model.

mixed_df = clean_data.dropna(subset=["Score"]).copy()

model_formula = "Score ~ C(Model) + C(Domain)"
lmm_model = smf.mixedlm(model_formula, mixed_df, groups=mixed_df["Evaluator_ID"])
lmm_result = lmm_model.fit(reml=False)

print(lmm_result.summary())
```

```python
print("Running fixed-effects OLS for omnibus significance check...")

ols_model = smf.ols("Score ~ C(Model) + C(Domain) + C(Evaluator_ID)", data=mixed_df).fit()
anova_table = sm.stats.anova_lm(ols_model, typ=2)
display(anova_table)
```

```python
print("Calculating agreement metrics...")

# Matrix: rows = subjects (Raw_Column), columns = raters (Evaluator_ID)
irr_df = clean_data[["Raw_Column", "Evaluator_ID", "Score"]].dropna().copy()
irr_matrix_df = irr_df.pivot(index="Raw_Column", columns="Evaluator_ID", values="Score")

# Exact agreement: proportion of rows where all raters gave same score
def exact_agreement(row):
    vals = row.dropna().values
    if len(vals) <= 1:
        return np.nan
    return 1.0 if np.all(vals == vals[0]) else 0.0

# Relaxed agreement: max difference <= tolerance
def relaxed_agreement(row, tolerance=1):
    vals = row.dropna().values
    if len(vals) <= 1:
        return np.nan
    return 1.0 if (np.max(vals) - np.min(vals) <= tolerance) else 0.0

exact_scores = irr_matrix_df.apply(exact_agreement, axis=1)
relaxed_scores = irr_matrix_df.apply(lambda r: relaxed_agreement(r, tolerance=1), axis=1)

print(f"Absolute % Agreement (Exact): {100 * np.nanmean(exact_scores):.2f}%")
print(f"Relaxed % Agreement (Within 1 point): {100 * np.nanmean(relaxed_scores):.2f}%")

# Pairwise weighted Cohen's kappa across evaluators, then average
raters = irr_matrix_df.columns.tolist()
pairwise_kappas = []

for r1, r2 in combinations(raters, 2):
    temp = irr_matrix_df[[r1, r2]].dropna()
    if len(temp) > 0:
        kappa = cohen_kappa_score(temp[r1], temp[r2], weights="quadratic")
        pairwise_kappas.append(kappa)

if pairwise_kappas:
    print(f"Mean Pairwise Weighted Cohen's Kappa: {np.mean(pairwise_kappas):.4f}")
    print(f"Median Pairwise Weighted Cohen's Kappa: {np.median(pairwise_kappas):.4f}")
else:
    print("Not enough overlapping ratings to compute kappa.")
```
